## Agentic AI and Frameworks

- `AI Agent` - LLM model that performs a specific task e.g. writing emails. It has an LLM in the background that is integrated with tools built by a developer.
- If we connect/integrate multiple AI agents we get `Agentic AI.` Agentic AI is autonomous systems that perceive, reason, plan and act independently to perform complex tasks with minimal human intervention.
- Put it simply, AI agents perform a subgoal, agentic AI performs the overall goal.
- `Framework` - same as a module e.g. reusable script that has functions, methods etc.
